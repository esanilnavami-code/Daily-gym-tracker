```python
df = pd.read_csv('/kaggle/input/daily-gym-tracker/gym.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Converting to datetime datatype
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['date_year'] = df['Date'].dt.year

df.sample(5)
```

```python
df['date_month'] = df['Date'].dt.month

df.sample(5)
```

```python
df['date_day'] = df['Date'].dt.day

df.sample(5)
```

```python
df['date_month_name'] = df['Date'].dt.month_name()

df.head()
```

```python
df['date_day_of_w'] = df['Date'].dt.dayofweek

df.head()
```

```python
df['date_dow_name'] = df['Date'].dt.day_name()

df.sample(10)
```

```python
df['date_is_weekend'] = np.where(df['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

df.sample(10)
```

```python
df['date_quarter'] = df['Date'].dt.quarter

df.sample(10)
```

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - df['Date']
```

```python
(today - df['Date']).dt.days
```

```python
print(df[df['time'] == '0'])
```

```python
print(df.dtypes)
```

```python
df['time'] = df['time'].astype(str)
```

```python
df.dtypes
```

```python
df['time'] = df['time'].replace('0', '00:00:00')

df.head()
```

```python
df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S', errors='coerce').dt.time
```

```python
df
```

```python
df['hour'] = df['time'].apply(lambda t: t.hour)
df['minute'] = df['time'].apply(lambda t: t.minute)
df['second'] = df['time'].apply(lambda t: t.second)
```
